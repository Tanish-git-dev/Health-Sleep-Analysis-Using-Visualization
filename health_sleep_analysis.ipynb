```python
#Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Reading the data to make the dataset
df = pd.read_csv("health.csv")
#First 5 rows of the data
df.head()
```

```python
#Making a scatter plot with regresstion line between Stress Level and Quality of Sleep to find out the correlation
plt.figure(figsize=(8,5))
sns.regplot(
    data=df,
    x="Stress Level",
    y="Quality of Sleep"
)
plt.title("Stress Level vs Quality of Sleep")
plt.xlabel("Stress Level")
plt.ylabel("Quality of Sleep")
plt.grid(True)
plt.show()
```

```python
#Making a scatter plot with regresslion line between Physical Activity Level and Quality of Sleep to find the correlation
plt.figure(figsize=(8,5))
sns.regplot(
    data = df,
    x = "Physical Activity Level",
    y = "Quality of Sleep"
)
plt.title("Physical Activity Level vs Quality of Sleep")
plt.xlabel("Physical Activity Level")
plt.ylabel("Quality of Sleep")
plt.grid(True)
plt.show()
```

```python
#Making a scatter plot with regresstion line between Age and Quality of Sleep to find out the correlation
plt.figure(figsize=(8,5))
sns.regplot(
    data=df,
    x="Age",
    y="Quality of Sleep"
)
plt.title("Age vs Quality of Sleep")
plt.xlabel("Age")
plt.ylabel("Quality of Sleep")
plt.grid(True)
plt.show()
```

```python
#Making a box plots between Sleep Quality and number of REM cycles completed
df["REM Cycles"] = (df["Sleep Duration"] / 1.5).astype(int) 
#Dividing by 1.5 since a REM cycle is of 90 minutes on average
#Converting it to int since number of REM cycles can't be a decimal value

plt.figure(figsize=(8,5))
sns.boxplot(
    data = df,
    x = "REM Cycles",
    y = "Quality of Sleep"
)
plt.title("REM Cycles vs Sleep Quality")
plt.xlabel("REM Cycles completed")
plt.ylabel("Quality of Sleep")
plt.grid(True)
plt.show()
```

```python
#Making a bar graph between BMI Category and probability of having sleep disorder
df['Sleep_Problem'] = df['Sleep Disorder'].notna().astype(int) # No Disorder = 0, Disorder = 1

plt.figure(figsize=(8,5))
sns.barplot(
    data=df.assign(**{'BMI Category': df['BMI Category'].replace({'Normal': 'Normal Weight'})}),
    x='BMI Category',
    y='Sleep_Problem',
    order=['Normal Weight', 'Overweight', 'Obese']
)
plt.grid(True)
plt.show()
```

```python
#Making a box plot to find out difference between sleeping hours of people with disorders to no disorder
df["Sleep Disorder"] = df["Sleep Disorder"].fillna("No Disorder") #Filling null values with No Disorder

plt.figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x="Sleep Disorder",
    y="Sleep Duration",
    order=["No Disorder", "Sleep Apnea", "Insomnia"]
)
plt.title("Sleep Duration by Sleep Disorder")
plt.xlabel("Sleep Disorder")
plt.ylabel("Sleep Duration (hours)")
plt.grid(True)
plt.show()
```
